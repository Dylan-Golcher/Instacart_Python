```python
#Limit the dataframe to the first million rows
df = df_ords_prods_merged[:1000000]
```

```python
#Create user defined condition
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#Define the string function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
#Check the subset using the price_range function
df['price_range'].value_counts(dropna = False)
```

```python
#Check the max value of the subset
df['prices'].max()
```

```python
#Define the string function using loc
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
#Check the value counts of string function using loc
df['price_range'].value_counts(dropna = False)
```

```python
#Define the string function using loc of the whole dataframe
df_ords_prods_merged.loc[df_ords_prods_merged['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
#Check which days of the week have highest/lowest order counts
df_ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
#Use for loop to create a result set
result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
#Combine the for loop with the dataframe
df_ords_prods_merged['Busiest day'] = result
```

There are 24 hours of orders in a day and three time periods requested for the analysis (fewest, average, most). I took 24 and divided by 3, so the top 8 are most, the middle 8 are average, and the bottom 8 are fewest.

```python
#Define result3 set
result3 = []

for value in df_ords_prods_merged["order_hour_of_day"]:
  if value in [9,10,11,12,13,14,15,16]:
    result3.append("Most orders")
  elif value in [7,8,17,18,19,20,21,22]:
    result3.append("Average orders")
  elif value in [0,1,2,3,4,5,6,23]:
    result3.append("Fewest orders")
```

```python
#Merge result3 with the dataframe to create new column
df_ords_prods_merged['busiest_period_of_day'] = result3
```
