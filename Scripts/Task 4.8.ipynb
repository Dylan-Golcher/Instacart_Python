## Question 6

```python
#Create a new column 'spending_rank' for the aggregated values
df_ords_prods_merged['spending_rank'] = df_ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#Rename column for clarity
df_ords_prods_merged.rename(columns = {'spending_rank':'avg_price'}, inplace = True)
```

```python
#Create the flag
df_ords_prods_merged.loc[df_ords_prods_merged['avg_price'] >= 10, 'spending_flag'] = 'High spender'
df_ords_prods_merged.loc[df_ords_prods_merged['avg_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
#Check the new column
df_ords_prods_merged.head(10)
```

```python
#Check the value counts
df_ords_prods_merged['spending_flag'].value_counts(dropna = False)
```

## Question 7

```python
#Create a new column for the order frequency
df_ords_prods_merged['median_days_since_prior_order'] = df_ords_prods_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
df_ords_prods_merged.head(5)
```

```python
#Rename column to be shorter
df_ords_prods_merged.rename(columns = {'median_days_since_prior_order':'median_dspo'}, inplace = True)
```

```python
df_ords_prods_merged.head(5)
```

```python
#Create the flag
df_ords_prods_merged.loc[df_ords_prods_merged['median_dspo'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
df_ords_prods_merged.loc[(df_ords_prods_merged['median_dspo'] > 10) & (df_ords_prods_merged['median_dspo'] <= 20), 'order_frequency_flag'] = 'Regular customer'
df_ords_prods_merged.loc[df_ords_prods_merged['median_dspo'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
df_ords_prods_merged.head(5)
```

```python
#Check the value counts
df_ords_prods_merged['order_frequency_flag'].value_counts(dropna = False)
```
